```python
import pandas as pd
```

```python
import nltk 
import numpy as np
```

```python
filename = '../NLP_LAB4/spam.csv'
df = pd.read_csv(filename,encoding = 'latin-1')
#df = pd.read_csv(filename,encoding = 'utf-8')
print(df.shape)
print(df.columns)
print(df.head(2))
```

```python
data = df.values
print(data.shape)
```

```python
print(data[:2])
```

```python
lines = data[:,1]
labels = data[:,0]
print(lines[:5])
print(labels[:20])
```

```python
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
def remove_stop_words(words):
    result = [i for i in words if i not in stopwords.words('english')]
    return result
```

```python
# tách từ trong từng câu 
sentences = [nltk.word_tokenize(sent) for sent in lines]
print(sentences[0])
remove_stop_words(sentences)
# thêm 2 gram 
```

```python
from keras.preprocessing.text import Tokenizer 
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
# kiểm tra từ điểm tạo trong tokenizer
#print(tokenizer.word_index)
```

```python
#texts_to_matrix(tokenizer, texts, mode = c("binary", "count", "tfidf", "freq"))
X = tokenizer.texts_to_matrix(sentences,mode='binary')
print(X)
```

```python
# chia tập train và tập test 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression  
model_LR = LogisticRegression()
model_LR = model_LR.fit(X_train,y_train)
y_pred = model_LR.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_pred))
```

```python
from sklearn.naive_bayes import MultinomialNB   
model_NB = MultinomialNB()
model_NB = model_NB.fit(X_train,y_train)
y_pred = model_NB.predict(X_test)

print(classification_report(y_test,y_pred))
```
